#### Merging the 12 months of sales data into a single csv file 

```python
path = "./Sales_Data"
files = [file for file in os.listdir('Sales_Data') if not file.startswith('.')]    # Ignore hidden files

all_months_data = pd.DataFrame()          # empty data frame 

for file in files : 
    current_data = pd.read_csv(path+"/"+file)
    all_months_data = pd.concat([all_months_data, current_data])
  
all_months_data.to_csv("all_data_copy.csv", index=False) 
```

```python
all_data = pd.read_csv("all_data_copy.csv")
all_data.head(10)
```

### dataframe copy

```python
all_data_copy = all_data.copy()
```

### Clean up the data : Drop rows of NAN

```python
all_data_copy.dropna(subset = ['Order ID'],inplace=True)
all_data_copy.head(10)
```

### Adding month column

```python
all_data_copy = all_data_copy[all_data_copy['Quantity Ordered'] != 'Quantity Ordered']
all_data_copy['Order Date'] = pd.to_datetime(all_data_copy['Order Date'])
all_data_copy['Month'] = all_data_copy['Order Date'].dt.month
all_data_copy.reset_index(drop=True,inplace=True)
all_data_copy.head(10)
```

### Make columns correct type

```python
all_data_copy['Price Each'] = pd.to_numeric(all_data_copy['Price Each'])
all_data_copy['Quantity Ordered'] = pd.to_numeric(all_data_copy['Quantity Ordered'])  
all_data_copy.dtypes                                             
```

# Data Exploration
### Question 1: What was the best month for sales? How much was earned that month?

```python
all_data_copy['Sales'] = all_data_copy['Quantity Ordered'] * all_data_copy['Price Each']
all_data_copy.groupby(['Month']).sum()
```

```python
months = all_data_copy.groupby('Month').sum().reset_index()['Month'].unique()
plt.figure(figsize=(15, 6))
plt.bar(months,all_data_copy.groupby(['Month']).sum().Sales)
plt.title('What was the best month for sales?')
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month')
plt.show()
```

### Question 2: What city sold the most product?

```python
def get_city(adress) : 
    return adress.split(',')[1].strip('')
def get_state(address):
    return address.split(",")[2].split(" ")[1]
```

### Adding City column

```python
all_data_copy['City'] = all_data_copy['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
all_data_copy.head(10)
```

```python
all_data_copy.groupby('City').sum()
```

```python
all_data_copy.groupby('City').sum().reset_index()
```

```python
all_data_copy.groupby('City').sum().reset_index()['City'].unique()
```

```python
cities = all_data_copy.groupby('City').sum().reset_index()['City'].unique()

plt.figure(figsize=(15, 6))
plt.bar(cities,all_data_copy.groupby(['City']).sum().Sales)
plt.title('What city sold the most product?')
plt.xticks(cities, rotation='45')
plt.ylabel('Sales in USD ($)')
plt.xlabel('City')
plt.show()
```

### Question 4: What products are most often sold together?

### keep the duplicated Order ID

```python
df = all_data_copy[all_data_copy['Order ID'].duplicated(keep=False)]
df.head(10)
```

```python
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df = df[['Order ID','Grouped']].drop_duplicates()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
products_list = []
products_count = []

for key,value in count.most_common(10):
    products_list.append(key)
    products_count.append(value)

products_list = [','.join(elem) for elem in products_list ]
```

```python
plt.figure(figsize=(15, 6))
plt.bar(products_list,products_count, color='g')
plt.title('What products are most often sold together?')
plt.xticks(products_list, rotation='75' , size='12')
plt.ylabel('number of time')
plt.xlabel('product couples')
plt.show()
```
